# Plot Function to see Model Performance

```python
def plot_function(history):
    fig,ax = plt.subplots(1,2,figsize = (25,5))
    ax[0].plot(history.history['loss'],color = 'red',label = 'train_loss')
    ax[0].set_title('Loss and val_loss')
    ax[0].plot(history.history['val_loss'],color = 'green',label = 'val_loss')
    ax[0].legend()
    ax[1].plot(history.history['accuracy'],color = 'orange',label = 'train_accuracy')
    ax[1].set_title('accuracy and val_accuracy')
    ax[1].plot(history.history['val_accuracy'],color = 'black',label = 'val_accuarcy')
    ax[1].legend()
```

# Trying out different Neural Network Architectures

1. LeNet5 inspired model

```python
def LeNet(input_shape,num_labels):
    Inputs = Input(shape = input_shape)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(Inputs)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = AveragePooling2D(pool_size = (2,2),strides = 1)(x)
    x = AveragePooling2D(pool_size = (2,2),strides = 2)(x)
    x = Flatten()(x)
    x = Dense(120,activation = 'relu')(x)
    x = Dense(84,activation = 'relu')(x)
    Outputs = Dense(num_labels,activation = 'softmax')(x)
    LeNet_model = Model(Inputs,Outputs)
    return LeNet_model
```

```python
input_shape = (image_size,image_size,3)
model = LeNet(input_shape = input_shape,num_labels = num_labels)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001,decay = 6e-5)
model.compile(loss = 'binary_crossentropy',optimizer = optimizer,metrics = ['accuracy'])
model.summary()
```

```python
history1 = model.fit(x_train,y_train,batch_size = 10,epochs = 100,validation_data = (x_valid,y_valid))
```

```python
plot_function(history1)
```

```python
loss,acc = model.evaluate(x_test,y_test,batch_size = 10)
print('Loss:',loss,'Accuracy: ',acc)
```

2. Adding Dropout to the same model

```python
def LeNet1(input_shape,num_labels):
    Inputs = Input(shape = input_shape)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(Inputs)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Dropout(0.3)(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Dropout(0.3)(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Dropout(0.3)(x)
    x = Conv2D(filters = 32,kernel_size = (5,5),strides = 1,padding = 'same',activation = 'relu')(x)
    x = Dropout(0.3)(x)
    x = AveragePooling2D(pool_size = (2,2),strides = 1)(x)
    x = Dropout(0.3)(x)
    x = AveragePooling2D(pool_size = (2,2),strides = 2)(x)
    x = Flatten()(x)
    x = Dense(120,activation = 'relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(84,activation = 'relu')(x)
    Outputs = Dense(num_labels,activation = 'softmax')(x)
    LeNet_model1 = Model(Inputs,Outputs)
    return LeNet_model1
```

```python
input_shape = (image_size,image_size,3)
model1 = LeNet1(input_shape = input_shape,num_labels = num_labels)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001,decay = 6e-5)
model1.compile(loss = 'binary_crossentropy',optimizer = optimizer,metrics = ['accuracy'])
model1.summary()
```

```python
history2 = model1.fit(x_train,y_train,batch_size = 10,epochs = 100,validation_data = (x_valid,y_valid))
```

3. Y Network

```python
def ynetwork(input_shape,num_labels):
    input1 = Input(shape = input_shape)
    x = input1
    x = Conv2D(32,kernel_size = (2,2),padding = 'same',activation = 'relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.3)(x)
    x = Conv2D(64,kernel_size = (2,2),padding = 'same',activation = 'relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.3)(x)
    x = Conv2D(64,kernel_size = (2,2),padding = 'same',activation = 'relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.4)(x)
    x = Conv2D(128,kernel_size = (2,2),padding = 'same',activation = 'relu')(x)
    x = Dropout(0.3)(x)
 

    input2 = Input(shape = input_shape)
    y = input2
    y = Conv2D(32,kernel_size = (2,2),padding = 'same',activation = 'relu')(y)
    y = MaxPooling2D(pool_size = (2,2))(y)
    y = Dropout(0.3)(y)
    y = Conv2D(64,kernel_size = (2,2),padding = 'same',activation = 'relu')(y)
    y = MaxPooling2D(pool_size = (2,2))(y)
    y = Dropout(0.3)(y)
    y = Conv2D(64,kernel_size = (2,2),padding = 'same',activation = 'relu')(y)
    y = MaxPooling2D(pool_size = (2,2))(y)
    y = Dropout(0.3)(y)
    y = Conv2D(128,kernel_size = (2,2),padding = 'same',activation = 'relu')(y)
    y = Dropout(0.3)(y)
    
    z = concatenate([x,y])
    z = Flatten()(z)
    z = Dense(256,activation = 'relu')(z)
    z = Dense(64,activation = 'relu')(z)
    outputs = Dense(num_labels,activation = 'sigmoid')(z)
    y_net = Model([input1,input2],outputs)
    return y_net
```

```python
model2 = ynetwork(input_shape = input_shape,num_labels = num_labels)
model2.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
history3 = model2.fit([x_train,x_train],y_train,batch_size =10,epochs = 100,validation_data = ([x_valid,x_valid],y_valid))
```

```python
model2.evaluate([x_test,x_test],y_test,batch_size = 10)
```
